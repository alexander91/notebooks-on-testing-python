```python
from scipy.stats.distributions import norm
```

$$\sigma = \sqrt {\frac{\hbar \tau}{m}}$$

```python
sigma = 1.0
```

```python
def perform_random_step(walkers):
    step = numpy.random.normal(0, 1.0, len(walkers))
    step *= sigma
    for i in range(len(walkers)):
        walkers[i] += step[i]
```

```python
walkers = pylab.arange(-0.1, 0.1, 0.000001)
for i in range(1000):
    if ((i % 100) == 0):
        matplotlib.pyplot.hist(walkers, 50, normed = True, range=(-100, 100))
        matplotlib.pyplot.ylim(0, 0.15)
        matplotlib.pyplot.show()
    perform_random_step(walkers)
```
